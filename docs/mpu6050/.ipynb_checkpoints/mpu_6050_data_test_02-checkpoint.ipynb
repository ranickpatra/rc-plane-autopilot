```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Calibration data

```python
columns = ['gx', 'gy', 'gz', 'ax', 'ay', 'az', 'temp']

with open('./static.txt', 'r') as f:
    lines = f.readlines()
    lines = [[float(_x.strip()) for _x in x.split(',')] for x in lines]
    data_static = pd.DataFrame(data=lines, columns=columns)
    
calibration_g = np.array([
    np.average(data_static['gx'].values),
    np.average(data_static['gy'].values),
    np.average(data_static['gz'].values)
])

calibration_g
```

# Data Movement

```python
columns = ['gx', 'gy', 'gz', 'ax', 'ay', 'az', 'temp']
sampling_rate = 250

with open('./movement.txt', 'r') as f:
    lines = f.readlines()
    lines = [[float(_x.strip()) for _x in x.split(',')] for x in lines]
    data_movement = pd.DataFrame(data=lines, columns=columns)
    
# calibration
data_movement = data_movement - np.array([calibration_g[0], calibration_g[1], calibration_g[2],
                                         0, 0, 0, 0])
```

```python
# Calculate angle from gyro only

data_angle_rate_g = data_movement[['gx', 'gy', 'gz']] / (65.5 * sampling_rate)
data_angle_g = []
for row in data_angle_rate_g.values:
    if len(data_angle_g) == 0:
        data_angle_g.append(row)
    else:
        data_angle_g.append(data_angle_g[-1] + row)

data_angle_g = np.array(data_angle_g)
data_angle_g = pd.DataFrame(data=data_angle_g, columns=['x', 'y', 'z'])

plt.figure(figsize=(20, 6))
plt.plot(data_angle_g['z'], label='z')
plt.legend(loc="upper left")
plt.show()
```

```python
# calculate angle from acclerrometer only

data_angle_a = []
for row in data_movement[['ax', 'ay', 'az']].values:
    row = row * 9.81 / 4096 # convert to m/s^2
    data_angle_a.append(np.arctan(row[0] / row[1]))
    
data_angle_a = np.array(data_angle_a)

plt.figure(figsize=(20, 6))
plt.plot(data_angle_a, label='z')
plt.legend(loc="upper left")
plt.show()
```

```python
# using complementry filter

data_angle_cmpl = []

for i in range(data_angle_rate_g.shape[0]):
    if len(data_angle_cmpl) == 0:
        data_angle_cmpl.append(data_angle_a[i])
    else:
        data_angle_cmpl.append((data_angle_cmpl[-1] + data_angle_rate_g.iloc[i][['gz']].values) * 0.96 + data_angle_a[i] * 0.04)

data_angle_cmpl = np.array(data_angle_cmpl)


plt.figure(figsize=(20, 6))
plt.plot(data_angle_cmpl, label='z')
plt.legend(loc="upper left")
plt.show()
```
