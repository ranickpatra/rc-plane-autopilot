```python
%matplotlib inline
from IPython.display import Markdown, display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
columns = ['angle', 'fin', 'motor']
data = []
with open('./data_2.txt', 'r') as f:
    data = [l.strip()[:-1] for l in f.readlines() if len(l.strip()) > 0]
data = [[float(_d) for _d in d.split(',')] for d in data]
data = np.array(data)
df_raw = pd.DataFrame(data=data, columns=columns)
df_raw
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
df_normalized = pd.DataFrame(scaler.fit_transform(df_raw), columns=df_raw.columns)

plt.figure(figsize=(20,7))
plt.plot(df_normalized[['angle', 'motor']])
plt.legend(df_normalized[['angle', 'motor']].columns)
plt.show()
```

```python
# plot sections
def plot_sections(df, ranges):
    df = df[ranges[0]:ranges[1]]
    plt.figure(figsize=(20,7))
    plt.plot(df)
    plt.legend(df.columns)
    plt.show()
```

```python
plot_sections(df_normalized, [600,1200])
```

```python
plot_sections(df_normalized, [2800,3500])
```

```python
plot_sections(df_normalized, [4800,5600])
```

```python
plot_sections(df_normalized, [1000,2000])
```
