```python
%matplotlib inline
from IPython.display import Markdown, display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
columns = ['roll', 'pitch', 'yaw', 'prop', 'fin_1', 'fin_2', 'fin_3', 'fin_4']
with open('./flight_data_01.txt', 'r') as f:
    df_raw = [l.strip()[:-1] for l in f.readlines() if len(l.strip()) > 0]
df_raw = [[float(_d) for _d in d.split(',')] for d in df_raw]
df_raw = pd.DataFrame(data=np.array(df_raw), columns=columns)
df_raw
```

```python
skip = 2800
end = len(df_raw)
# end = 3100


plt.figure(figsize=(20,15))

columns_to_plot = ['prop']
plt.subplot(3,1,1)
plt.plot(df_raw[columns_to_plot][skip:end])
plt.legend(columns_to_plot)

columns_to_plot = ['roll', 'pitch', 'yaw']
plt.subplot(3,1,2)
plt.plot(df_raw[columns_to_plot][skip:end])
plt.legend(columns_to_plot)

plt.subplot(3,1,3)
# columns_to_plot = ['fin_1', 'fin_2', 'fin_3', 'fin_4']
columns_to_plot = ['fin_1', 'fin_4']
plt.plot(df_raw[columns_to_plot][skip:end])
plt.legend(columns_to_plot)

# Adjust layout to prevent overlap
plt.tight_layout()

plt.show()
```
