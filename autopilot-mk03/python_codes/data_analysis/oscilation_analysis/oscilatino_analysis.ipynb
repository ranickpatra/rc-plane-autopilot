```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
delta_t = 0.01
```

```python
columns = ['roll', 'pitch', 'yaw', 'rps', 'fin1', 'fin2', 'fin3', 'fin4']
data = []
with open('./data_20240714_135554.txt', 'r') as f:
    data = [l.strip()[:-1] for l in f.readlines() if len(l.strip()) > 0]
data = [[float(_d) for _d in d.split(',')] for d in data]
data = np.array(data)
df_raw_data = pd.DataFrame(data=data, columns=columns)
df_raw_data
```

```python
plt.figure(figsize=(20,20))

columns_to_plot = ['rps']
plt.subplot(6,1,1)
plt.plot(df_raw_data[columns_to_plot])
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)

columns_to_plot = ['roll', 'pitch', 'yaw']
plt.subplot(6,1,2)
plt.plot(df_raw_data[columns_to_plot])
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)




columns_to_plot = ['fin1']
plt.subplot(6,1,3)
plt.plot(df_raw_data[columns_to_plot], color='red')
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)

columns_to_plot = ['fin3']
plt.subplot(6,1,4)
plt.plot(df_raw_data[columns_to_plot], color='green')
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)

columns_to_plot = ['fin2']
plt.subplot(6,1,5)
plt.plot(df_raw_data[columns_to_plot], color='indigo')
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)

columns_to_plot = ['fin4']
plt.subplot(6,1,6)
plt.plot(df_raw_data[columns_to_plot], color='orange')
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)




# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,15))

df = df_raw_data[750:1600]

columns_to_plot = ['rps']
plt.subplot(2,1,1)
plt.plot(df[columns_to_plot])
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)

columns_to_plot = ['roll', 'fin1']
plt.subplot(2,1,2)
plt.plot(df[columns_to_plot])
plt.legend(columns_to_plot)
plt.grid(which='both', linewidth=0.5)


# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
